```python
# Regula-Falsi Method

import numpy as np
import math
import pylab

def f(x):
    return 112.5 * x**2 - 234.84375 * x + 81.81875

# Interval 0.1 <= x <= 0.9
a = 0.1
b = 0.9
maxer = 0.01
maxit = 100
fa = f(a)
fb = f(b)
x1=[]
y1=[]

def regula_falsi(f, a, b, maxit, maxer):
    n = 1
    print('\t || \t iteration \t ||\t x \t\t||\t f(x) \t\t||')

    while n<=maxit:
        c = ((f(a)*b)-(f(b)*a))/(f(a)-f(b))
        fc = f(c)
        print('\t || \t ', n, '\t\t ||\t', (round(c, 5)), '\t||\t', (round(fc, 5)), '\t||')

        if f(a) * f(c) < 0:
            b = c
        else:
            a = c

        x1.append(n)
        y1.append(c)

        if abs(fc)<=maxer:
          break

        n += 1

    if n > maxit:
        print("Algorithm fails: no convergence")

#Call Regula-Falsi method function
regula_falsi(f, a, b, maxit, maxer)

pylab.plot(x1,y1,'o-')
```
