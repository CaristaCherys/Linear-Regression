```python
# Gauss Elimination

import numpy as np
import pandas as pd
import pylab
# Input ordo matrix
R = []
X1 = []
X2 = []

# Array
R_values = []
X1_values = []
X2_values = []

# Loop
for r in np.arange(0.6, 1.0, 0.05):

    a = np.array([[0.5 * r, 1.5 * r - 3.237], [18.409, -0.5]])
    b = np.array([[-0.7], [2.818]])
    n = len(a)
    print('------------------------------------------')
    print('R = ', round(r, 2))

    print('Matrix a:')
    print(a)

    print('Matrix b:')
    print(b)

    Matrix = np.array(a, float)
    Hasil = np.array(b, float)

    for k in range(0, n - 1):
        for i in range(k + 1, n):
            if Matrix[i, k] != 0:
                x = Matrix[i, k] / Matrix[k, k]
                Matrix[i, k:n+1] = Matrix[i, k:n+1] - Matrix[k, k:n+1] * x
                Hasil[i] = Hasil[i] - (Hasil[k] * x)
    print("Matrix setelah eliminasi : ", '\n', Matrix)
    print("Hasil setelah eliminasi : ", '\n', Hasil)

    # Substitusi komponen matrix bernilai 0
    X = np.zeros(n, float)
    for m in range(n - 1, -1, -1):
        X[m] = (Hasil[m] - np.dot(Matrix[m, m+1:n], X[m+1:n])) / Matrix[m, m]
        print('Nilai X',m + 1, '=', X[m])

    # Tambah hasil pada list
    R_values.append(round(r, 2))
    X1_values.append(X[0])
    X2_values.append(X[1])

# Menampilkan data dalam tabel menggunakan fungsi pandas
df = pd.DataFrame({'R': R_values, 'X1': X1_values, 'X2': X2_values})
print('\nResults Table:')
print(df)

# Menampilkan plot grafik R terhadap X1 dan X2
print('PLOT GRAFIK R TERHADAP X1 DAN X2')
pylab.plot(R_values,X1_values)
pylab.plot(R_values,X2_values)
```
