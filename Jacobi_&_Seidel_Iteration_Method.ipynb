```python
# JACOBI ITERATION

import numpy as np
from numpy import *
import pylab
from prettytable import PrettyTable

t = PrettyTable() # Tabel hasil iterasi Jacobi untuk setiap r
t1 = PrettyTable() # Tabel A dan B serta hasil SPL untuk setiap r
t2 = PrettyTable() # Tabel keseluruhan hasil iterasi Jacobi X1 dan X2 setiap iterasi

t2.field_names = ['R','X1','X2']

# Menampilkan tebakan awal
x = np.array([[0.0],[0.0]])
print ('Tebakan awal = \n', x)

x1final = []
x2final = []
R = []
x1 = []
x2 = []
kx = []

# Looping setiap r dari 0,6 hingga 1 dengan interval 0,05
for r in arange (0.6,1.0,0.05):
  t.field_names = ['Iterasi','X1','Error X1','X2','Error X2']
  t1.field_names = ['A','B','Hasil SPL']

  print ('\nR = ',(round(r,2)))

  A = np.array([[18.409,-0.5],[0.5*r,(1.5*r-3.237)]])
  B = np.array([[2.818],[-0.7]])
  n = len(A)
  xbaru = zeros ((n,1),float)
  error = zeros ((n,1),float)
  maxit = 20
  k = 0

  # Looping iterasi Jacobi
  while (k<=maxit):
    for i in range (0,n):
      s = 0
      for j in range (0,n):
        if (j!=i):
          s = s+(A[i,j]*x[j])
      xbaru[i] = (B[i]-s) /A[i,i]
      error[i] = abs((xbaru[i]-x[i])/xbaru[i])
      x[:] = xbaru[:]
    x1.append (xbaru[0,0])
    x2.append (xbaru[1,0])
    kx.append(k)
    t.add_row([k,(round(xbaru[0,0],5)),error[0,0], round(xbaru[1,0],5), error[1,0]])
    k+=1

  x1final.append(xbaru[0,0])
  x2final.append(xbaru[1,0])
  R.append(r)

  # Menampilkan tabel 1 dan tabel 2
  t1.add_row([A,B,xbaru])
  t2.add_row([round(r,2),xbaru[0,0],xbaru[1,0]])
  t.align = 'c'
  t1.align = 'c'
  print (t1)
  print (t)
  t.clear()
  t1.clear()

# Menampilkan hasil pada tabel dan grafik
print ('\n\t TABEL HASIL ITERASI JACOBI \t\t')
print (t2)

print ('\n\t GRAFIK HASIL ITERASI JACOBI \t\t')
pylab.plot (R,x1final,'-o')
pylab.plot (R,x2final,'-x')
```
